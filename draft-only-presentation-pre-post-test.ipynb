## Pre-Test and Post-Test Design

#### **Pre-Test Design**

1. **Define Outcomes:** Identify what you’re measuring to form the hypothesis for analysis (e.g., skills, knowledge).

2. **Choose Format:** Select the test method (quiz, survey) based on the type of data that will influence your statistical test.

3. **Pilot:** Test on a small group to ensure accuracy and refine questions before full administration.

4. **Administer Consistently:** Ensure uniform conditions during the test to avoid bias and allow for valid statistical significance testing later.

#### **Post-Test Design**

1. **Match with Pre-Test:** Measure the same outcomes as the pre-test to allow accurate comparison.

2. **Maintain Conditions:** Keep the testing environment and instructions consistent to make results valid for significance analysis.

3. **Use Similar Questions:** Ensure questions are comparable to the pre-test for valid statistical calculations.

4. **Test Shortly After Intervention:** Administer the test soon after the intervention to accurately capture its effects for interpretation.

# Step 1: Defining Hypotheses

- #### **Null Hypothesis (H₀):** No difference between pre-test and post-test results.
- #### **Alternative Hypothesis (H₁):** Significant difference between pre-test and post-test results.

# Step 2: Choosing the Right Statistical Test

- #### a: **Paired t-test:** Use this if your data is continuous, normally distributed, and you have a paired design (same individuals tested before and after).
- #### b: **Wilcoxon Signed-Rank Test:** Use this if your data is continuous or ordinal but not normally distributed.
- #### c: **McNemar’s Test:** For paired nominal data (e.g., success/failure).

#### Use the **p-value** from your chosen test to assess whether the observed differences are statistically significant. If the p-value is less than your alpha level (e.g., 0.05), reject the null hypothesis.



##### Getting the p-value from the Paired t-test

1. Calculate the t-value:

2. Use the p-value chart or calculator:

	- You can use a calculator or software to find the p-value. For 3 people and a t-value of 3.46, the p-value is about 0.05.

##### What Does the p-value Tell Us?
- If the p-value is less than 0.05, it means the results are statistically significant (the changes are unlikely to be by chance).
