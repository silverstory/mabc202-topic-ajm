### **What is a Pre-Test?**

- A **pre-test** is an evaluation given to participants before they undergo an intervention. It provides a baseline for their initial knowledge or state.
- For example, in a classroom, pre-tests help assess the initial level of knowledge or skills. A pre-test on algebra can show how much students know before starting the subject, allowing comparison to the post-test results to assess changes.

---

### **What is a Post-Test?**

- A **post-test** is an evaluation given to participants after completing an intervention to assess the changes or improvements.
- The post-test helps determine the effectiveness of the intervention by comparing results with the pre-test.  
- **Example (Student Context):**  
  After completing the math course, students take a post-test to evaluate what theyâ€™ve learned, and comparing the scores reveals their progress.

---

### **Why Compare Pre-Test and Post-Test?**

- Comparing pretest and posttest scores helps determine the effectiveness of an intervention or program.
- This comparison shows whether the participants have improved, indicating the success of the intervention.  
- It reveals if the intervention or program was effective by measuring the improvement.
- **Example (Student Context):**  
  In a school, comparing pre-test and post-test results from a science lesson can reveal how much students learned during the class.

### Why subtract 1 from N? \(N - 1\)

The reason we use \(N - 1\) instead of just \(N\) when calculating the **sample standard deviation** is due to something called **Bessel's correction**. 

### Explanation:
1. **Sample vs. Population:**
   - When we are working with a **sample** of data (a subset of a larger population), we are trying to estimate the variability (spread) of the entire population based on that small sample.
   - If we only divide by \(N\), we might **underestimate** the variability in the population. This is because a small sample is likely to be less spread out than the whole population.

2. **Besselâ€™s Correction:**
   - To correct for this bias, we subtract **1** from \(N\). This correction is known as **Bessel's correction**, and it makes the standard deviation a **better estimate** of the population standard deviation.

3. **Degrees of Freedom:**
   - The subtraction of **1** corresponds to the idea of **degrees of freedom**. When calculating the mean, we use up one degree of freedom because the mean depends on the sample data, leaving us with \(N - 1\) independent data points. Therefore, we adjust the divisor to reflect this.

### Intuition:
By subtracting **1**, we're allowing for a slightly larger standard deviation to compensate for the fact that we are working with a sample and not the entire population. Without this correction, the estimate of the population standard deviation would tend to be too small.

In summary, the "1" accounts for the fact that the mean is an estimate from the sample, and we want to avoid underestimating the true variability in the population when using sample data.

### **T-Test Formula:**
- Used to compare the means of two sets of data, like pre-test and post-test scores, to check if the difference is significant.

  

  

### One-sample, two-sample, or paired t test?

- If the groups come from a single population (e.g., measuring before and after an experimental treatment), perform a paired t test. This is a [within-subjects design]().
   - In a within-subjects design, or a within-groups design, all participants take part in every condition. Itâ€™s the opposite of a between-subjects design, where each participant experiences only one condition.
   - A within-subjects design is also called a dependent groups or repeated measures design because researchers compare related measures from the same participants between different conditions.

- If the groups come from two different populations (e.g., two different species, or people from two separate cities), perform a two-sample t test (a.k.a. independent t test). This is a between-subjects design.

- If there is one group being compared against a standard value (e.g., comparing the acidity of a liquid to a neutral pH of 7), perform a one-sample t test.

#### One-tailed or two-tailed t test?

- If you only care whether the two populations are different from one another, perform a two-tailed t test.
- If you want to know whether one population mean is greater than or less than the other, perform a one-tailed t test.

### **Calculating Pre-Test and Post-Test Results**

#### **1. Generate Sample Data using Python:**

```python
# Simulated dataset for pretest and posttest scores
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


pd.options.mode.chained_assignment = None  # default='warn'

# Generating Random Sample Data
np.random.seed(0)
student_ids = np.arange(1, 31)
pre_test_scores = np.random.randint(40, 80, 30)
post_test_scores = pre_test_scores + np.random.randint(-5, 15, 30)

# Creating a DataFrame
df = pd.DataFrame({'Student-Id': student_ids, 'Pre-Test': pre_test_scores, 'Post-Test': post_test_scores})

# Display the Data
print(df)
```

---

#### **2. Calculate the Mean and Standard Deviation:**

```python
# Calculating Mean
pre_mean = df['Pre-Test'].mean()
post_mean = df['Post-Test'].mean()

# Calculating Standard Deviation
pre_std = df['Pre-Test'].std()
post_std = df['Post-Test'].std()

print(f"Pre-Test Mean: {pre_mean}, Post-Test Mean: {post_mean}")
print(f"Pre-Test Standard Deviation: {pre_std}, Post-Test Standard Deviation: {post_std}")
```

---

#### **3. Perform a t-Test to Determine Significance:**

```python
from scipy import stats

# Paired t-test
t_stat, p_value = stats.ttest_rel(df['Pre-Test'], df['Post-Test'])

print(f"t-Statistic: {t_stat}, p-value: {p_value}")
```

**Results:**
- The mean, standard deviation, and t-test results allow you to analyze whether the difference in scores is statistically significant.
- If the p-value is less than 0.05, the result is significant, indicating that the post-test improvement is not due to chance.

---

### **Scatter Plot Interpretation of Scores**

**1. Visualize the Data:**

```python
# Scatter plot of pretest and posttest scores
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Pre-Test', y='Post-Test', data=df, hue='Student-Id', palette='coolwarm')
plt.title('Scatter Plot of Pretest vs Posttest Scores')
plt.show()
```

```python
# Mean Calculation
pre_mean = np.mean(df['Pre-Test'])
post_mean = np.mean(df['Post-Test'])

# Standard Deviation Calculation using ddof=1 for sample standard deviation
pre_std = np.std(df['Pre-Test'], ddof=1) # degree of freedom = 1
post_std = np.std(df['Post-Test'], ddof=1)

# Degrees of freedom are the number of independent variables that can be estimated
# in a statistical analysis and tell you how many items can be randomly selected
# before constraints must be put in place.

# t-Test Calculation
t_stat, p_val = stats.ttest_rel(df['Pre-Test'], df['Post-Test'])

# Display Results
print(f"Pre-Test Mean: {pre_mean}, Post-Test Mean: {post_mean}")
print(f"Pre-Test Std Dev: {pre_std:.2f}, Post-Test Std Dev: {post_std:.2f}")
print(f"t-Statistic: {t_stat}, p-value: {p_val}")

# Visualization with Matplotlib and Seaborn
plt.figure(figsize=(10,5))

# Visualize the distribution of scores
# plt.figure()
sns.kdeplot(df['Pre-Test'], label="Pre-Test", fill=True)
sns.kdeplot(df['Post-Test'], label="Post-Test", fill=True)
plt.title("Distribution of Pre-Test and Post-Test Scores")
plt.xlabel("Score")
plt.ylabel("Density")
plt.legend()

plt.show()
```

**2. Interpretation of Results (Significance and Effect Size):**

- **Significant Results:**  
  If the t-test shows that the post-test scores are significantly higher (p < 0.05), it means that the intervention had a meaningful impact.
  
- **Non-Significant Results:**  
  If p > 0.05, the intervention might not have been as effective, and the change in scores could be due to chance.

- **Effect Size:**  
  To calculate the effect size and understand the magnitude of improvement, you can compute Cohenâ€™s d:

```python
# Calculating Cohen's d
mean_diff = post_mean - pre_mean
pooled_std = np.sqrt((pre_std**2 + post_std**2) / 2)
cohens_d = mean_diff / pooled_std

print(f"Cohen's d: {cohens_d}")
```

Cohen's d. The standardised difference between two means.

- Is a measure of effect size that indicates the standardized difference between two means. The formula for Cohenâ€™s d is:





- It can be used, for example, to accompany reporting of t-test and ANOVA results. It is also widely used in meta-analysis. Cohen's d is an appropriate effect size for the comparison between two means.

### **Types of Pretest-Posttest Designs**

#### **1. Randomized Control Group Pretest-Posttest**:
- In this design, participants are **randomly assigned** to two groups:
   - **Treatment Group**: Receives the intervention.
   - **Control Group**: Does not receive the intervention.
- Both groups take a **pretest** and a **posttest**.
- This design helps establish whether the treatment caused any changes by comparing the difference in the scores from pretest to posttest between the two groups.

#### **2. Nonrandomized Control Group Pretest-Posttest**:
- Similar to the above design but **without random assignment**.
- Pretest and posttest scores are still used for both groups, but since the groups are not randomly assigned, there might be **bias**.

#### **3. Randomized Solomon Four-Group Design**:
- **Four groups** are used:
   1. **Pretest + Treatment**: This group takes a pretest, receives intervention, then takes a posttest.
   2. **Pretest + Control**: This group takes a pretest, does not receive intervention, then takes a posttest.
   3. **No Pretest + Treatment**: This group skips the pretest, receives intervention, and then takes a posttest.
   4. **No Pretest + Control**: This group skips the pretest, does not receive intervention, and then takes a posttest.
   
- **Purpose**: This design allows researchers to see whether taking the pretest itself has an effect on the results (a testing effect). 


---

#### Example illustration of results from Randomized Solomon Four-Group design

```python
# Simulate data for the four-group design
group_data = {
    'Group': ['Pretest + Treatment', 'Pretest + Control', 'No Pretest + Treatment', 'No Pretest + Control'],
    'Pretest_Score': [np.nan, 75, np.nan, 70],
    'Posttest_Score': [85, 78, 88, 72]
}

# Convert the data into a DataFrame
group_df = pd.DataFrame(group_data)

# Create a heatmap for comparison of group scores
plt.figure(figsize=(8, 4))
sns.heatmap(group_df.set_index('Group'), annot=True, cmap='Blues', linewidths=0.5)
plt.title('Randomized Solomon Four-Group Design Scores')
plt.show()
```

### **What does this heatmap show?**:

- **Heatmap Rows**: Each row corresponds to one of the four groups (Pretest + Treatment, Pretest + Control, No Pretest + Treatment, No Pretest + Control).
  
- **Pretest and Posttest Scores**: The heatmap shows the scores in each stage (pretest and posttest):

- **Blank Values**: Groups that didn't take the pretest (like "No Pretest + Treatment" and "No Pretest + Control") show **Blank** in the heatmap for the pretest column. This visually emphasizes that those groups skipped the pretest phase.

- **Color Gradient**: The intensity of the colors represents the scores, with darker blues representing higher scores and lighter blues representing lower scores.

- "This design helps to determine if the pretest itself has any influence on the posttest results. By comparing the groups that took the pretest with those that didn't, we can see whether the act of taking the pretest affects learning or performance."

### **Limitations of Pre-Test and Post-Test Comparisons**

- **Limitations:**  
  Results can be affected by external factors like participant motivation, external study, or fatigue.
- Some students may perform worse on the post-test due to distractions or stress, not necessarily because they didnâ€™t learn from the intervention.
- **Example:**  
  In a classroom, students who didnâ€™t sleep well the night before the post-test may score lower, even if they learned from the lesson.

---

### **Pros and Cons of Pretest-Posttest**

**Pros:**
- **Clear Comparison Between Before and After:**
  - "One of the major advantages is that it allows for a **clear comparison** of participants' performance before and after the intervention. This can provide insightful data on how effective the treatment has been."
  
- **Good for Measuring Change Over Time:**
  - "Additionally, this design is particularly useful for **measuring change over time**. It enables researchers to see how much improvement has occurred, making it valuable for evaluating educational programs, therapies, or any interventions."


**Cons:**
- **Testing Effect:**
  - "However, there are some drawbacks. One significant issue is the **testing effect**. If participants are familiar with the test format or content, their scores may improve simply due to practice rather than the effectiveness of the treatment."
  
- **May Not Account for All Variables:**
  - "Another limitation is that pretest-posttest designs might not account for all confounding variables. Factors outside the treatment could influence the results, leading to inaccurate conclusions."
