- ### How to Determine a Pre-Test

A pre-test measures participants’ baseline knowledge, skills, or behaviors before an intervention to assess progress later.

- Define the Outcomes: Focus on what you want to measure (e.g., knowledge, skills).
- Align with Program Goals: Ensure the pre-test content matches the objectives of your intervention.
- Choose the Format: Use quizzes, surveys, or performance tests depending on what you're evaluating.
- Pilot the Test: Test with a small group to ensure clarity.
- Administer the Pre-Test: Ensure consistent conditions (e.g., time, environment) to avoid bias.

Use descriptive statistics or a paired t-test to summarize results and compare with the post-test.

- ### How to Determine a Post-Test

A post-test measures the impact of the intervention by comparing results with the pre-test.

- Align with Pre-Test: Ensure it assesses the same concepts and outcomes.
- Keep Conditions Consistent: Use the same environment and instructions as the pre-test.
- Use Comparable Questions: Modify pre-test questions slightly to avoid memorization.
- Administer Immediately: Conduct the post-test soon after the intervention (unless delayed follow-up is needed).

Use a paired t-test, gain scores, or Cohen’s d to quantify improvement and measure the intervention's effectiveness.

---

### To determine Pre-Test/Post-Test statistics, you generally compare the outcomes of a group before (pre-test) and after (post-test) an intervention or treatment.

# Slide 2: Defining Hypotheses

- #### **Null Hypothesis (H₀):** No difference between pre-test and post-test results.
- #### **Alternative Hypothesis (H₁):** Significant difference between pre-test and post-test results.

---


# Slide 3: Choosing the Right Statistical Test

- #### **Paired t-test:** Use this if your data is continuous, normally distributed, and you have a paired design (same individuals tested before and after).
- #### **Wilcoxon Signed-Rank Test:** Use this if your data is continuous or ordinal but not normally distributed.
- #### **McNemar’s Test:** For paired nominal data (e.g., success/failure).

---
