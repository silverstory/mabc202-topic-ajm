## Pre-Test and Post-Test Design

#### **Pre-Test Design**

1. **Define Outcomes:** Identify what you’re measuring to form the hypothesis for analysis (e.g., skills, knowledge).

2. **Choose Format:** Select the test method (quiz, survey) based on the type of data that will influence your statistical test.

3. **Pilot:** Test on a small group to ensure accuracy and refine questions before full administration.

4. **Administer Consistently:** Ensure uniform conditions during the test to avoid bias and allow for valid statistical significance testing later.

#### **Post-Test Design**

1. **Match with Pre-Test:** Measure the same outcomes as the pre-test to allow accurate comparison.

2. **Maintain Conditions:** Keep the testing environment and instructions consistent to make results valid for significance analysis.

3. **Use Similar Questions:** Ensure questions are comparable to the pre-test for valid statistical calculations.

4. **Test Shortly After Intervention:** Administer the test soon after the intervention to accurately capture its effects for interpretation.

# Step 2-b: Wilcoxon Signed-Rank Test

#### For a **Wilcoxon Signed-Rank Test**, rank the differences in **"absolute value"**, keeping track of the signs, and then use these ranks to compute the test statistic.

### 1. Pre-test and Post-test Scores Example

**Data:**
- Pre-test scores: 70, 75, 80
- Post-test scores: 85, 80, 90

#### Step 1: Calculate the Differences
First, subtract the pre-test score from the post-test score for each person.

| Subject | Pre-test | Post-test | Difference (Post - Pre) |
|---------|----------|-----------|-------------------------|
| 1       | 70       | 85        | 15                      |
| 2       | 75       | 80        | 5                       |
| 3       | 80       | 90        | 10                      |

#### Step 2: Rank the Absolute Differences
Next, rank the absolute differences (ignoring the signs for now):

- Absolute differences: |15|, |5|, |10|
- Ranks: The smallest absolute difference (5) gets Rank 1, the next (10) gets Rank 2, and the largest (15) gets Rank 3.

| Subject | Difference | Absolute Difference | Rank |
|---------|------------|---------------------|------|
| 1       | 15         | 15                  | 3    |
| 2       | 5          | 5                   | 1    |
| 3       | 10         | 10                  | 2    |

#### Step 3: Apply the Signs to the Ranks
Now, reintroduce the original signs of the differences:

| Subject | Difference | Rank | Signed Rank |
|---------|------------|------|-------------|
| 1       | +15        | 3    | +3          |
| 2       | +5         | 1    | +1          |
| 3       | +10        | 2    | +2          |

#### Step 4: Compute the Test Statistic
The test statistic $$ W $$ is the sum of the signed ranks:

$$
W = (+3) + (+1) + (+2) = 6
$$

So, the test statistic for the pre-test and post-test scores is **6**.

---


To test if the **W** statistic in the **Wilcoxon Signed Rank Test** is significant;

### Option 1. **Find the Critical Value**

   You can compare the **(W)** value against critical values from a **Wilcoxon Signed Rank Test** table (found in statistics textbooks) for the given sample size **(n)** and significance level (**alpha**, e.g., 0.05). The steps are:
   
   - Look up the critical value for your **sample size (n)** and **significance level** (e.g., (**alpha** = 0.05)).
   - If the absolute value of **W** is **less than or equal to** the critical value, reject the null hypothesis (**H₀**).
   - If the absolute value of **W** is **greater than** the critical value, fail to reject the null hypothesis.




### Wilcoxon Signed-Rank Test:
- We calculate differences, rank absolute values, apply signs, and sum the signed ranks to get the test statistic.
- **W critical method**: Compare **W** with the critical value from the table.

# Step 2-c: McNemar’s Test

### What is McNemar’s Test?

McNemar’s Test is used when you want to check if there’s a significant change between **two paired, binary outcomes**. Typically, it's applied when measuring **before and after** results on the same group of people.

- **Paired**: You are measuring the same people twice (before and after).
- **Binary outcomes**: There are only two possible outcomes (e.g., **Yes/No**, **Pass/Fail**, **Improved/Not Improved**).

It’s often used in medical or educational studies where you want to see if a treatment or intervention made a difference.

### Example Scenario:

Let’s say a researcher wants to know if a new medication improves patients’ conditions. The patients are tested **before** and **after** taking the medication. The outcomes are binary: each patient either **gets better** (yes) or **does not get better** (no).

We want to know if there is a significant change in the patients’ health outcomes **before** and **after** the treatment using McNemar's Test.

---

### Why Use a Contingency Table?

McNemar’s Test uses a **2x2 contingency table** because we are comparing **two paired binary outcomes** (before and after) for each individual.

Here’s what each part of the table represents:

|                   | **After: Did Not Improve** | **After: Improved** |
|-------------------|----------------------------|---------------------|
| **Before: Did Not Improve**  | a (No → No)           | b (No → Yes)         |
| **Before: Improved**         | c (Yes → No)          | d (Yes → Yes)        |

- **a**: People who did not improve before, and still did not improve after.
- **b**: People who did not improve before, but improved after.
- **c**: People who improved before, but did not improve after.
- **d**: People who improved before and still improved after.

The key part of McNemar’s Test focuses on **b** and **c**, which are the mismatched pairs (those who changed outcomes). These tell us whether the number of people who improved after failing initially is significantly different from those who worsened after succeeding initially.

---

### Example

Example of a class taking a test before and after studying with a new method:

- There are 10 students.
- Each student either **passes** or **fails** both the **pre-test** and the **post-test**.

After the new study method, the results are recorded like this:

|                   | **Post-test: Failed** | **Post-test: Passed** |
|-------------------|-----------------------|-----------------------|
| **Pre-test: Failed**  | 2 students (a)        | 4 students (b)         |
| **Pre-test: Passed**  | 3 students (c)        | 1 student (d)          |

- **a = 2**: Students who failed both the pre-test and the post-test.
- **b = 4**: Students who failed the pre-test but passed the post-test.
- **c = 3**: Students who passed the pre-test but failed the post-test.
- **d = 1**: Students who passed both the pre-test and the post-test.

### 4. How to Calculate McNemar’s Test Statistic

McNemar’s Test focuses on the off-diagonal elements \( b \) and \( c \) because these represent **changes in outcome**. The test checks if the number of people who **improved (b)** is significantly different from the number who **got worse (c)**.

The formula for McNemar’s Test is:

$$
\chi^2 = \frac{(b - c)^2}{b + c}
$$

Where:
- \( b \) = 4 (students who improved: failed pre-test, passed post-test)
- \( c \) = 3 (students who got worse: passed pre-test, failed post-test)

Now, plug in the values:

$$
\chi^2 = \frac{(4 - 3)^2}{4 + 3}
$$

$$
\chi^2 = \frac{1^2}{7} = \frac{1}{7} \approx 0.14
$$

### 5. Interpreting the Test Statistic

The test statistic $$\chi^2$$ is **0.14**. 

To know if this is significant, we would compare it to a critical value from the chi-square distribution (with 1 degree of freedom). For most studies, the critical value at a 0.05 significance level is 3.84.



Since 0.14 is much smaller than 3.84, we would **not** reject the null hypothesis. This means there is **no significant difference** in the number of students who improved vs. those who got worse.

---

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the chi-square value from the McNemar's test (students' test scores example)
chi_square_value = 0.14  # Chi-square value calculated from the students' data

# Define the critical value for alpha = 0.05
alpha = 0.05
df = 1  # degrees of freedom (McNemar's test always uses df=1)
critical_value = chi2.ppf(1 - alpha, df)

# Create a range of x values (chi-square values)
x = np.linspace(0, 15, 500)

# Plot the chi-square distribution
plt.figure(figsize=(8, 6))
plt.plot(x, chi2.pdf(x, df), color='blue', lw=2, label='Chi-Square Distribution (df=1)')

# Shade the critical region (x > critical_value)
plt.fill_between(x, 0, chi2.pdf(x, df), where=(x > critical_value), color='red', alpha=0.5, label=fr'Critical Region ($\chi^2 > {critical_value:.2f}$)')

# Mark the chi-square value from McNemar's test (students' example)
plt.axvline(chi_square_value, color='green', linestyle='--', lw=2, label=fr'$\chi^2 = {chi_square_value}$')

# Mark the critical value
plt.axvline(critical_value, color='black', linestyle='--', lw=2, label=f'Critical Value = {critical_value:.2f}')

# Add labels and title
plt.title(f'Chi-Square Distribution (df=1) with Critical Region for $\\alpha={alpha}$')
plt.xlabel(r'$\chi^2$ Value')
plt.ylabel('Probability Density')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

### Explanation of the Graph:

- The **blue curve** shows the chi-square distribution with **1 degree of freedom**.
- The **red-shaded region** is the critical region where χ² values exceed the critical value (3.84) at a **significance level of α = 0.05**. Any χ² value falling here would lead us to reject the null hypothesis.
- The **green dashed line** represents the χ² value from McNemar's test in this example, which is **0.14**.
- Since the green line is **not** in the red region, we do **not** reject the null hypothesis. This means there is no statistically significant difference between the pre-test and post-test results for the students.

### Key Points to Understand McNemar’s Test:
- McNemar’s test is useful when studying **before and after** changes in the same group of people.
- It focuses on **binary outcomes** (e.g., Pass/Fail, Yes/No).
- The contingency table summarizes paired outcomes for each person (before and after).
- The test checks if there is a significant difference between the **number of people who changed in one direction** (e.g., got better) versus those who changed in the opposite direction (e.g., got worse).

## Example: **one-tailed, right-tailed test**.

### Hypotheses Setup:
- **Null Hypothesis (H₀)**: Post-test scores are **not higher** than pre-test scores i.e., $$ \mu_{\text{post}} \leq \mu_{\text{pre}} $$
- **Alternative Hypothesis (H₁)**: Post-test scores are **higher** than pre-test scores i.e., $$ \mu_{\text{post}} > \mu_{\text{pre}} $$

### Tail Direction:
- **right-tailed test**: the post-test scores are significantly higher than pre-test scores.
- A **right-tailed test** focuses on the upper end of the distribution, where higher scores would reject the null hypothesis in favor of the alternative.

### What Happens If the t-Stat is Negative?
- In a **right-tailed test**, a **negative t-value** (like our **t Stat = -3.464**) suggests that post-test scores are **lower** than pre-test scores.
- Since a **negative t-value** is far from the positive critical region e.g., $$ t_{critical} = 2.919 $$, we **fail to reject the null hypothesis**. In other words, the data does not support the claim that post-test scores are higher than pre-test scores.

- Given our **t Stat = -3.464**, the result would **not be statistically significant** in a **right-tailed test** because:
- The t-value is negative, meaning post-test scores are likely lower than pre-test scores.
- It falls outside the positive critical region e.g., $$ t_{critical} = 2.919 $$
- Thus, we **fail to reject the null hypothesis** and conclude that **post-test scores are not higher** (and might even be lower) than the pre-test scores.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Data
before = np.array([70, 75, 80])
after = np.array([85, 80, 90])

# Create the DataFrame for Before and After
data_df = pd.DataFrame({
    'Before': before,
    'After': after
})

# Calculating statistics
mean_before = np.mean(before)
mean_after = np.mean(after)
variance_before = np.var(before, ddof=1)
variance_after = np.var(after, ddof=1)
observations = len(before)
pearson_corr = np.corrcoef(before, after)[0, 1]
t_stat, p_value_two_tailed = stats.ttest_rel(before, after)
df = observations - 1
p_value_one_tailed = p_value_two_tailed / 2

# Manually setting the critical t-value for the right-tailed test
t_critical_one_tail = 2.919

# Plotting the t-distribution with the right-tailed critical region
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(10, 6))
sns.lineplot(x=x, y=y, color='blue')

# Shade the right-tail critical region (positive)
x_fill_right = np.linspace(t_critical_one_tail, 4, 500)
y_fill_right = stats.t.pdf(x_fill_right, df)
plt.fill_between(x_fill_right, y_fill_right, color='maroon', alpha=0.6)

# Mark the t-stat value on the plot
plt.axvline(t_stat, color='green', linestyle='--', label=f't Stat = {round(t_stat, 3)}')

# Annotate the t-critical value and alpha level
plt.text(t_critical_one_tail + 0.2, 0.02, 'α = 0.05', fontsize=12, ha='center')
plt.axvline(t_critical_one_tail, color='black', linestyle='--', 
            label=f't Critical (right) = {round(t_critical_one_tail, 3)}')

# Labels and title
plt.title(f'T-Distribution Plot (Right-Tailed Test)\nT, df={df}, Critical t = {t_critical_one_tail}', fontsize=16)
plt.xlabel('T-Value', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.legend()

# Show the plot
plt.show()
```

## Indicator 2: Measuring Effect Size

### To measure the magnitude of the difference, compute the effect size. Common measures are:
- #### **Cohen’s d (for t-test)**
- #### **r (for non-parametric tests like Wilcoxon)**





### Calculate Cohen's d:



- <i>Cohen’s d is a way to measure how big the difference really is. In this case, it’s like asking, "Is this change a lot or just a little?"</i>

---

```python
import numpy as np
import pandas as pd

# Scores
before = np.array([70, 75, 80])
after = np.array([85, 80, 90])

# Calculate means and standard deviations
mean_before = np.mean(before)
mean_after = np.mean(after)
std_before = np.std(before, ddof=1)

# Calculate Cohen's d and convert to native Python float
cohens_d = float(round((mean_after - mean_before) / std_before, 2))

# Compare effect size
if abs(cohens_d) < 0.2:
    effect_size = "small"
elif abs(cohens_d) < 0.5:
    effect_size = "medium"
else:
    effect_size = "large"

# Output
cohens_d, effect_size
```



#### d = 2, which means the effect size is huge (the scores improved a lot!).
